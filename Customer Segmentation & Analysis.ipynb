```python
sns.pairplot(data)
```

```python
import seaborn as sns
sns.scatterplot(data=data,x='Annual_Income_(k$)', y='Spending_Score',color='orange', s=150, alpha=0.4)
plt.title('K-means Customer Segments')
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score')
plt.show()
```

```python
sns.violinplot(data=data,x='Annual_Income_(k$)',color='green')
```

```python
sns.kdeplot(data=data, x='Annual_Income_(k$)', hue='Genre',shade=True)
```

```python
dist = ['Age', 'Annual_Income_(k$)','Spending_Score']
for i in dist:
    plt.figure(figsize=(5,3))
    sns.distplot(data[i],bins=15)
```

```python
dist = ['Age', 'Annual_Income_(k$)','Spending_Score']
for i in dist:
    plt.figure()
    sns.kdeplot(data=data, x=i, shade=True, hue='Genre')
    plt.title(f'Distribution of {i} by Gender')
    plt.show()
```

```python
dist = ['Age', 'Annual_Income_(k$)','Spending_Score']
for i in dist:
    plt.figure()
    sns.boxplot(data=data, x="Genre", y=i,palette='pastel')
    plt.title(f'Distribution of {i} by Gender')
    plt.show()
```
